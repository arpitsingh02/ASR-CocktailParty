```python
import os
import museval
import librosa
import shutil
import numpy as np
import wave
# from utils import SNR 
def stft(data, fft_size=512, step_size=160,padding=True):
    # short time fourier transform
    if padding == True:
        # for 16K sample rate data, 48192-192 = 48000
        pad = np.zeros(192,)
        data = np.concatenate((data,pad),axis=0)
    # padding hanning window 512-400 = 112
    window = np.concatenate((np.zeros((56,)),np.hanning(fft_size-112),np.zeros((56,))),axis=0)
    win_num = (len(data) - fft_size) // step_size
    out = np.ndarray((win_num, fft_size), dtype=data.dtype)
    for i in range(win_num):
        left = int(i * step_size)
        right = int(left + fft_size)
        out[i] = data[left: right] * window
    F = np.fft.rfft(out, axis=1)
    return F

def real_imag_expand(c_data,dim='new'):
    # dim = 'new' or 'same'
    # expand the complex data to 2X data with true real and image number
    if dim == 'new':
        D = np.zeros((c_data.shape[0],c_data.shape[1],2))
        D[:,:,0] = np.real(c_data)
        D[:,:,1] = np.imag(c_data)
        return D
    if dim =='same':
        D = np.zeros((c_data.shape[0],c_data.shape[1]*2))
        D[:,::2] = np.real(c_data)
        D[:,1::2] = np.imag(c_data)
        return D

def fast_stft(data,power=False,**kwargs):
    # directly transform the wav to the input
    # power law = A**0.3 , to prevent loud audio from overwhelming soft audio
    if power:
        data = power_law(data)
    return real_imag_expand(stft(data))

def SNR(true_file,pred_file):
    T_true,_ = librosa.load(true_file,sr=16000)
    F_true = fast_stft(T_true)
    T_pred, _ = librosa.load(pred_file,sr=16000)
    F_pred = fast_stft(T_pred)
    F_inter = F_true - F_pred
    P_true = np.sum(np.square(F_true[:,:,0])+np.square(F_true[:,:,1]))
    P_inter = np.sum(np.square(F_inter[:,:,0])+np.square(F_inter[:,:,1]))
    return 10*np.log10(np.divide(P_true,P_inter))

d = './predc'
predc="./predc"
truec='./truec'
mixc='./mixc'
import pandas as pd
onlyFolders=[o for o in os.listdir(d) if os.path.isdir(os.path.join(d,o)) and o[0]!='.']
df = pd.DataFrame(columns = ['time', 'target', 'metric','score','track'])
index=0
modelList=[]
inputList=[]
for i in onlyFolders:
  print("At ",index+1," of 500")
  index+=1
  ts=os.path.join(truec,i)
  ps=os.path.join(predc,i)
  ms=os.path.join(mixc,i)
  mypath=ts
  onlyfiles = [f for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]
  for j in onlyfiles:
    result1=SNR(os.path.join(ts,j),os.path.join(ps,j))
    result2=SNR(os.path.join(ts,j),os.path.join(ms,j))
    modelList.append(result1)
    inputList.append(result2)
    # print(result1,result2)
    # df1=result.df
    # df=df.append(df1)
  # break
  # break
print("Average Output SNR= ",sum(modelList)/len(modelList))
print("Average Input SNR= ",sum(inputList)/len(inputList))
print("Max Output SNR= ",max(modelList))
print("Max Input SNR= ",max(inputList))
print("Min Output SNR= ",min(modelList))
print("Min Input SNR= ",min(inputList))

# print(len(onlyFolders))
# df.to_csv("Metrics_mix.csv")
```
