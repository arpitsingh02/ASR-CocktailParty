```python
import matplotlib.pyplot as plot
from scipy.io import wavfile
freq,signal=wavfile.read("/content/drive/MyDrive/some_output/sample3/speaker1.wav")
plot.subplot(211)

plot.title('Spectrogram of speaker 1')
plot.plot(signal)
plot.xlabel('Sample')
plot.ylabel('Amplitude')
plot.subplot(212)
plot.specgram(signal,Fs=freq)
plot.xlabel('Time')
plot.ylabel('Frequency')

 
```
